```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Declare 2 variables:

a = [1,2,3,4,5]
b = [6,7,8,9,10]
```

```python
# Plot graph using Seaborn:

sns.lineplot(x = a, y = b)
```

```python
# create dataframe:

import pandas as pd

df = pd.DataFrame({
                    "a" : a,
                    "b" : b}
                 )
df
```

```python
# plot graph with sns:
sns.lineplot(x =  a, y = b, data = df)
```

```python
# read new dataset:

df = pd.read_csv("penguins.csv")
df
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df)
```

```python
# with specific hue:

sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df, hue  = 'sex')
```

```python
# With specific colors:

sns.lineplot(x = 'bill_length_mm', 
             y = 'flipper_length_mm', 
             data = df, 
             hue  = 'sex', 
             palette = 'rocket_r')
```

```python
# Used markers to show the points:

sns.lineplot(x = 'bill_length_mm', 
             y = 'flipper_length_mm', 
             data = df, 
             hue  = 'sex', 
             palette = 'rocket_r',
             style = 'sex',
             markers = ["o", "<"])
```

```python
df1 = pd.read_csv("penguins.csv").head(20)

sns.lineplot(x = 'bill_length_mm', 
             y = 'flipper_length_mm', 
             data = df1, 
             hue  = 'sex', 
             palette = 'rocket_r',
             style = 'sex',
             markers = ["o", "<"])
plt.grid()
plt.title("Graph related to penguins")
```

#### Bar Graph:

```python
df1 = pd.read_csv("penguins.csv").head(20)

sns.barplot(x = 'bill_length_mm', 
             y = 'flipper_length_mm', 
             data = df1, 
             hue  = 'sex', 
             palette = 'rocket_r',
           )
plt.grid()
plt.title("Graph related to penguins")
```

```python
df1 = pd.read_csv("penguins.csv")

sns.barplot(x = df1.island, 
            y = df1.bill_length_mm,
            data = df1,
            hue = 'sex'
           )
plt.grid()
plt.legend()
plt.title("BarPlot")
plt.show()
```

#### we also can plot this with specific order as per our need:

```python
o = ["Biscoe", "Torgersen", "Dream"]

sns.barplot(
            x = "island",
            y = "bill_length_mm",
            data = df1, 
            hue = "sex",
            order = o
            )
plt.title("Bar Graph with specific order of Island") 
plt.legend()
```

#### Hue_order:

```python
o = ["Biscoe", "Torgersen", "Dream"]

sns.barplot(
            x = "island",
            y = "bill_length_mm",
            data = df1, 
            hue = 'sex',
            hue_order = ["FEMALE", "MALE"],
            order = o,
            palette = 'Accent'
            )
plt.title("Bar Graph with specific order of Island") 
plt.legend()
```

#### Saturation parameter:

```python
o = ["Biscoe", "Torgersen", "Dream"]

sns.barplot(
            x = "island",
            y = "bill_length_mm",
            data = df1, 
            hue = 'sex',
            hue_order = ["FEMALE", "MALE"],
            order = o,
            palette = 'Accent',
            saturation = 0.1   # value is in range of 0  to 1
            )
plt.title("Bar Graph with specific order of Island") 
plt.legend()
```

### Histogram:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1.head(10)
```

```python
sns.displot(df1["flipper_length_mm"])
plt.title("Histogram")
```

### With proper interval: Bins

```python
sns.displot(df1["flipper_length_mm"],
           bins = [170,180,190,200,210,220,230],
           color = 'blue')

plt.title("Histogram")
```

### KDE:

```python
sns.displot(df1["flipper_length_mm"],
           bins = [170,180,190,200,210,220,230], kde = True,
           color = 'black')

plt.title("Histogram")
```
