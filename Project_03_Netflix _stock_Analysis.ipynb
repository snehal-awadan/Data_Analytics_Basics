## Objectives:

1. Volume of stock traded
2. Netflix stock price: High, Open, Close
3. Netflix stock price: Day, Month, Year wise
4. Top 5 Dates with highest stock price
5. Top 5 dates with lowest stock price
6. High and Low values stock per period of time

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("NFLX.csv")
df.head()
```

```python
# set the figure:

sns.set(rc = 
        {
            'figure.figsize' :(10,5)
        }
)
```

```python
# take date as index:

# first convert date into proper format of datetime:
df['Date'] = pd.to_datetime(df['Date'])

# set date as index:
df = df.set_index('Date')

df.head()
```

## Objective No. 01
#### Volume of stock traded

```python
sns.lineplot(x = df.index,
             y = df['Volume'],
             label = 'Volume',
            )
plt.title("Volume of stock versus Time")
```

## Objective No. 02

#### Netflix stock price: High, Open, Close

```python
df.plot(y = ['Open', 'Close', 'High'],
       title = 'Netflix stock Price')
```

## Objective No. 03
#### Netflix stock price: Day, Month, Year wise

```python
# plot 3 different subplot for day, month and year
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (15,25))

# plot for day:
df.groupby(df.index.day).mean().plot(y ='Volume',
                                     ax = ax1,
                                     xlabel = 'Day',
                                     title = "Netflix stock price day wise")

# plot for month:
df.groupby(df.index.month).mean().plot(y ='Volume',
                                     ax = ax2,
                                     xlabel = 'Month',
                                     title = "Netflix stock price month wise")

# plot for year:
df.groupby(df.index.year).mean().plot(y ='Volume',
                                     ax = ax3,
                                     xlabel = 'Year',
                                      title = "Netflix stock price year wise")
```

## Objective No. 04
#### Top 5 Dates with highest stock price

```python
a = df.sort_values(by = "High", ascending=False).head(5)
a['High']
```

## Objective No. 05
#### Top 5 Dates with lowest stock price

```python
highest = df.sort_values(by= 'Low', ascending=True).head(5)
highest['Low']
```

## Objective No. 06
#### High and Low values stock per period of time

```python
# Plot 2 different graphs :
fig, axes = plt.subplots(nrows = 1,
                        ncols = 2)

fig.suptitle("High and Low values stock per period of time",fontsize = 15)

sns.lineplot(ax = axes[0],
             y = df['High'], 
             x = df.index, 
             color = 'green'
            )

sns.lineplot(ax = axes[1],
             y = df['Low'], 
             x = df.index, 
             color = 'red'
            )
```
