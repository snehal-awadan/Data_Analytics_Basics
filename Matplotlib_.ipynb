```python
import matplotlib.pyplot as plt
```

#### Linear Graph:

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.title("Linear Graph")
plt.xlabel("Products")
plt.ylabel("Month")
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.xlabel("Products")
plt.ylabel("Month")
plt.plot(x,y,c)
plt.show()
```

```python
# To show the marker (plotted point) :

x = [8,7,5,6]
y = [5,8,1,3]
plt.plot(x,y,marker = 'o')
plt.xlabel("Products")
plt.ylabel("Month")
plt.show()
```

#### Bar Graph:


```python
x = [8,7,5,6]
y = [5,8,1,3]

plt.title("Bar plot")
plt.bar(x,y)
plt.show()
```

```python
x = [8,7,5,6]
y = [5,8,1,3]
c = ['r', 'g', 'b', 'y']
plt.bar(x,y,color = c)
plt.show()
```

```python
# horizontal bar graph:

x = ['A', 'B', 'C', 'D']
y = [8,4,9,2]
plt.xlabel("Products")
plt.ylabel("Month")
c = ['r', 'g', 'b', 'y']
plt.barh(x, y, color = c)
plt.show()
```

```python
# With define width: (in vertical bar)

x = [8,7,5,6]
y = [5,8,1,3]
c = ['r', 'g', 'b', 'y']
plt.bar(x,y,color = c, width = 0.1)
plt.show()
```

```python
# With define Height: (in horizontal bar)

x = [8,7,5,6]
y = [5,8,1,3]
c = ['r', 'g', 'b', 'y']
plt.barh(x,y, color = c, height = 0.1)
plt.show()
```

#### Scatter Plot:

```python
x = [1,4,7,8,9,11,47]
y = [7,9,5,4,6,48,96]

plt.title("scatter plot")
plt.xlabel("Products")
plt.ylabel("Month")
c = ["red", "violet", "black", "magenta", "blue", "purple", "pink"]
plt.scatter(x,y, color = c, s = 100)
plt.show()
```
